```python
#data cleaning
data cleaning mean fixing bad data in your data set
bad data could be
* empty cells
*data in wrong format
*wrong data
*duplicates
```

```python
#emppty cells
empty cells can potentially give you a wrong result when you analyze data

#remove rows
onw way to deal with empty cells is to remove rows that contain empty cells
this is usually ok,since data set can be very big and removing a few rows will not hav a big impack on the result.
```

```python
#returning a new data framewith no empty cells
import pandas as pd
df=pd.read_csv('/content/data.csv')
print(df.to_string())
new_df=df.dropna()
print(new_df.to_string())
```

```python
#by defsult the dropna() method returns a new dataframe and will not change the original.
#if you want to change the original dataframe use the inplace=True argument:
import pandas as pd
df.dropna(inplace=True)
print(df.to_string())
#note: now,the dropna(inplace=True) will Not return the new data frame ,
#but it will remove all rows contaioning null values from the original dataframe
```

```python
'''#replace empty values
another way of dealing with empty cells is to insert a new value instead.
this way you do not have to delete entire rows just because of some empty cells.
the fillna() method allows to replace empty cells with a values:
example
Replace NUll values with the number 130:'''
import pandas as pd
df=pd.read_csv('/content/data.csv')
print(df.to_string())
df.fillna(130,inplace=True)
print(df.to_string())
```

```python
'''replace only for specified columns
the example above replaces all empty cells in the whole dataframe
to only replace empty values for one column,specify the column name for the dataframe:
Example
Replace Null values in calories column with the number 130:'''
import pandas as pd
df=pd.read_csv('/content/data.csv')
print(df.to_string())
df['Calories'].fillna(130,inplace=True)
print(df.to_string())
```

```python
'''Replace using mean,median ,mode
a common way to replace empty cells is to calculate the mean median mode values of the column
pandas uses the mean() medain() mode() methods to calculate the respective values for a specified column
example
calculate the mean and replace any empty values with it'''
import pandas as pd
df=pd.read_csv('/content/data.csv')
x=df['Calories'].mean()
x=df['Calories'].fillna(x,inplace=True)
print(df.to_string())
```

```python
'''pandas -  cleaning data of wrong format
data of wrong format
cells with data of wrong format can make it difficult or even impossible to analyze data.
to fix it you have two options:
remocve the rows, or convert all cells in column into the same format
convet into the correct format
in our data frame we have two cells with the wrong format ,
the date column should be a string that represents a date:
convert to date:'''
import pandas as pd
df=pd.read_csv('/content/data.csv')
df=['Date']=pd.to_datetime(df['Date'])
print(df.to_string())
```

```python
#remove  rows with a Null values in the "date" column:
df.dropna(subset=['Date'],inplace=True)
```

```python
'''wrong data
option1 :Replacing Values
to fix wrong values we can to reolace them with something else.'''
import pandas as pd
df=pd.read_csv('/content/data.csv')
print(df.to_string())
df.loc[7,'Duration']=45
```

```python
'''Replacee big values
loop through all values in the "duration" column.
if the values is higher than the 120 set it to 120'''
for x in  df.index:
  if df.loc[x,"Duration"]>120:
    df.loc[x,"Duration"]=120
```

```python
'''Removing Rows
Anothr way of handling wrong data is to remove the rows that containing wrong data.
this way you do not have to find out what to replace them with and there is a good chance you do not need them to do your analyses
example
Delete rows where "duration" is higher tha 120'''
for x in df.index:
  if df.loc[x,"Duration"]>120:
    df.drop(x,inplace=True)
print(df.to_string())
```

```python
'''pandas: Removing Duplicates
the duplicated() method returns a boolean values for each row:
returns true for ec=very rows that is a duplicate otherwise False:'''
print(df.duplicated())
```
